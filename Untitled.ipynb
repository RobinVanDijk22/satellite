```python
paths = earthaccess.download(results, "data")
```

```python
dataset = xr.open_dataset(paths[0])
dataset
```

```python
array = np.log10(dataset["chlor_a"])
array.attrs.update(
    {
        "units": f'log10({dataset["chlor_a"].attrs["units"]})',
    }
)
crs_proj = cartopy.crs.Robinson()
crs_data = cartopy.crs.PlateCarree()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(projection=crs_proj)
array.plot(x="lon", y="lat", cmap="jet", ax=ax, robust=True, transform=crs_data)
ax.coastlines()
ax.set_title(dataset.attrs["product_name"])
plt.show()
```

```python
print(array.dims)
print(array.coords)
```

```python
# Define bounds
lat_min, lat_max = -63, -67
lon_min, lon_max = 93, 104

# Subset the data
subset = array.sel(
    lat=slice(lat_min, lat_max),
    lon=slice(lon_min, lon_max)
)

# Plot the subset
fig = plt.figure(figsize=(8, 6))
ax = plt.axes(projection=cartopy.crs.PlateCarree())
subset.plot(
    x="lon", y="lat", cmap="jet", ax=ax, robust=True, transform=cartopy.crs.PlateCarree()
)
ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=cartopy.crs.PlateCarree())
ax.coastlines()
ax.set_title("Chlorophyll-a (Denman region)")
plt.show()
```
